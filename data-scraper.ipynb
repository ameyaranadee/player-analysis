```python
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', None)  # Display all columns
pd.set_option('display.max_rows', None)     # Display all rows
np.set_printoptions(threshold=np.inf)       # Display entire numpy arrays
```

```python
webpage_names = ['stats', 'keepers', 'keepersadv', 'shooting', 'passing', 'passing_types', 'gca', 'defense', 'possession',
                'playingtime', 'misc']
```

```python
url = 'https://fbref.com/en/comps/Big5/stats/players/Big-5-European-Leagues-Stats'
driver.get(url)
driver.implicitly_wait(10)
```

```python
html_content = driver.page_source

# Create a BeautifulSoup object from the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Read the table data using pandas
tables = pd.read_html(url)

# Select the first table (index 0) on the page
df = tables[0]

# Extract column names
column_names = df.columns.tolist()

print("Column Names:")
headers = []
for col_name in column_names:
    headers.append(col_name[1])

print(headers)

# Drop the last row (contains empty cells) and reset the index
df = df.iloc[:-1].reset_index(drop=True)

# Extract player data
rows = df.values.tolist()

print("\nPlayer Data:")
for player_data in rows:
    print(player_data)
```

```python
player_stats = pd.DataFrame(data=rows, columns=headers)
player_stats.head(10)
```

```python
# Drop the last column
player_stats = player_stats.drop(player_stats.columns[-1], axis=1)
player_stats.head()
```
